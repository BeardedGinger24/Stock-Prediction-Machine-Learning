```python
import requests
import json
import pandas as pd

url = "https://newscatcher.p.rapidapi.com/v1/stocks"

page_number = 1

querystring = {"ticker":"TSLA",
               "from":"2016/03/26",
               "to":"2021/03/26",
               "lang":"en",
               "page": page_number,
               "media":"False",
               "sort_by":"date"}

headers = {
    'x-rapidapi-key': "0c21336d43msh23b54589bb91b35p1e18c9jsndba57676fb69",
    'x-rapidapi-host': "newscatcher.p.rapidapi.com"
    }

response = requests.request("GET", url, headers=headers, params=querystring)
response = json.loads(response.text)

total_pages = response['total_pages']
# print(response.text)
```

```python
articles = response['articles']

for page_number in range (2, total_pages + 1):
    querystring = {"ticker":"TSLA",
               "from":"2016/03/26",
               "to":"2021/03/26",
               "lang":"en",
               "page": page_number,
               "media":"False",
               "sort_by":"date"}
    
    response = requests.request("GET", url, headers=headers, params=querystring)
    response = json.loads(response.text)
    
    articles.extend(response['articles'])

```
